```python
import json
from datetime import datetime

ARQUIVO = "reservas.json"

class SistemaReserva:

    def __init__(self):
        self.reservas = self.carregar_dados()
        self.usuario_logado = None

    def carregar_dados(self):
        try:
            with open(ARQUIVO, "r") as f:
                return json.load(f)
        except:
            return []

    def salvar_dados(self):
        with open(ARQUIVO, "w") as f:
            json.dump(self.reservas, f, indent=4)

    # LOGIN
    def login(self, nome, usuario, senha):
        self.usuario_logado = {"nome": nome, "usuario": usuario, "senha": senha}
        print(f"✅ Bem-vindo(a), {nome}!")

    # VALIDAÇÃO DE DATA
    def validar_data(self, data):
        try:
            datetime.strptime(data, "%d/%m/%Y")
            return True
        except:
            return False

    # RESERVAR
    def reservar_lab(self, lab, unidade, data, horario):

        if not self.usuario_logado:
            print("❌ Faça login primeiro!")
            return

        if not self.validar_data(data):
            print("❌ Data inválida!")
            return

        # Verificar conflito
        for r in self.reservas:
            if all(k in r for k in ["lab", "unidade", "data", "horario"]):
                if (
                    r["lab"] == lab and
                    r["unidade"] == unidade and
                    r["data"] == data and
                    r["horario"] == horario
                ):
                    print("❌ Laboratório já reservado nesse dia e horário!")
                    return

        reserva = {
            "nome": self.usuario_logado["nome"],
            "usuario": self.usuario_logado["usuario"],
            "lab": lab,
            "unidade": unidade,
            "data": data,
            "horario": horario
        }

        self.reservas.append(reserva)
        self.salvar_dados()
        print("✅ Reserva realizada!")

    # LISTAR RESERVAS
    def ver_minhas_reservas(self):

        if not self.usuario_logado:
            print("❌ Faça login primeiro!")
            return

        print("\n📋 Suas reservas:")

        encontrou = False
        for r in self.reservas:
            if r.get("usuario") == self.usuario_logado["usuario"]:
                print(f"Lab: {r.get('lab', 'N/A')} | Unidade: {r.get('unidade', 'N/A')} | Data: {r.get('data', 'N/A')} | Horário: {r.get('horario', 'N/A')}")
                encontrou = True

        if not encontrou:
            print("Nenhuma reserva encontrada.")

    # CANCELAR RESERVA ESPECÍFICA
    def cancelar_reserva(self):

        if not self.usuario_logado:
            print("❌ Faça login primeiro!")
            return

        usuario = self.usuario_logado["usuario"]

        minhas_reservas = [r for r in self.reservas if r.get("usuario") == usuario]

        if not minhas_reservas:
            print("❌ Nenhuma reserva encontrada.")
            return

        print("\n🗑️ Suas reservas:")

        for i, r in enumerate(minhas_reservas, start=1):
            print(f"{i} - Lab: {r.get('lab', 'N/A')} | Unidade: {r.get('unidade', 'N/A')} | Data: {r.get('data', 'N/A')} | Horário: {r.get('horario', 'N/A')}")

        try:
            escolha = int(input("\nDigite o número da reserva que deseja cancelar: "))

            if escolha < 1 or escolha > len(minhas_reservas):
                print("❌ Opção inválida!")
                return

            reserva_escolhida = minhas_reservas[escolha - 1]

            confirmacao = input("⚠️ Tem certeza que deseja cancelar essa reserva? (s/n): ").lower()

            if confirmacao != "s":
                print("❌ Cancelamento abortado.")
                return

            self.reservas.remove(reserva_escolhida)
            self.salvar_dados()

            print("✅ Reserva cancelada com sucesso!")

        except ValueError:
            print("❌ Entrada inválida! Digite um número.")


def menu():
    sistema = SistemaReserva()

    while True:
        print("\n===== MENU =====")
        print("1 - Login")
        print("2 - Reservar lab")
        print("3 - Minhas reservas")
        print("4 - Cancelar reserva")
        print("0 - Sair")

        op = input("Escolha: ")

        if op == "1":
            nome = input("Nome: ")
            usuario = input("Usuario: ")
            senha = input("Senha: ")
            sistema.login(nome, usuario, senha)

        elif op == "2":
            lab = input("Lab: ")
            unidade = input("Unidade: ")
            data = input("Data (DD/MM/AAAA): ")
            horario = input("Horário (HH:MM): ")
            sistema.reservar_lab(lab, unidade, data, horario)

        elif op == "3":
            sistema.ver_minhas_reservas()

        elif op == "4":
            sistema.cancelar_reserva()

        elif op == "0":
            print("Encerrando...")
            break

        else:
            print("❌ Opção inválida!")


menu()
```
